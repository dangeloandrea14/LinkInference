```python
import pandas as pd
import os
from io import StringIO
from pandas import json_normalize
```

```python
config_file = os.path.join("output","runs","LinkAttack","Citeseer_SGC_edge.json")
with open(config_file, "r") as f:
    text = f.read().strip()

if not text.startswith("["):
    text = "[" + text
if text.endswith(","):
    text = text[:-1]
if not text.endswith("]"):
    text = text + "]"

df = pd.read_json(StringIO(text))

df_expanded = json_normalize(df['parameters'])
df = pd.concat([df.drop(columns='parameters'), df_expanded], axis=1)

cols = list(df.columns)

cols.insert(2, cols.pop(cols.index('optimizer.parameters.lr')))

df = df[cols[:22]]


display(df)
```
